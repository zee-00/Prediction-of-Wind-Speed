```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\zkyaf\\Downloads\\Met_data_1.csv")
```

```python
df.head()
```

```python
d=['RAIN_HOURLY(mm)','RAIN_DAILY(mm)','TEMP(C)','TEMP_MIN(C)','TEMP_MIN_MAX_DAY(C)','DEW_POINT(C)','RH(%)','RH_MIN_MAX_DAY(%)','WIND_DIR_10m(deg)','SLP(hPa)','MSLP(hPa)','SUNSHINE(HH:MM)','Unnamed: 16']
df.drop(d,axis=1,inplace=True)
```

```python
df['WIND_MAX/GUST_10m(km/h)']=df['WIND_MAX/GUST_10m(kt)']*1.852
```

```python
df.head()
```

```python
d=['WIND_SPEED_10m(kt)','WIND_MAX/GUST_10m(kt)']
df.drop(d,axis=1,inplace=True)
```

```python
df.head()
```

```python
df['TIME']=df['DATE'] +" "+ df['TIME(UTC)']
```

```python
df.head(3)
```

```python
d=['DATE','TIME(UTC)']
df.drop(d,axis=1,inplace=True)
```

```python
df.head()
```

```python
df2=df.set_index(pd.DatetimeIndex(df['TIME'])).drop('TIME',axis=1)
```

```python
df2.head(3)
```

```python
df3=df2.resample('3H').max()
```

```python
df3['WIND_MAX_GUST']=np.where(df3['WIND_MAX/GUST_10m(km/h)']>50,1,0)
```

```python
df3.head()
```

### 

```python
df3['Time']=df3.index
```

```python
df3['Month']=df3.index
```

```python
df3['Time']=df3['Time'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S')[11:])
```

```python
df3.sort_values(['Month','Time'])
```

```python
df3['Month']=df3['Month'].dt.strftime('%B')
```

```python
df3.head()
```

```python
df3=df3.reset_index(drop=True)
```

```python
df3.head()
```

```python
df3.to_csv('final_meta_data.csv')
```
